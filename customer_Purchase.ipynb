```python
# Calculate the average purchase amount for each city and product category
city_product_purchase = df.groupby(['City', 'Product_Category'])['Purchase_Amount'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(9, 6))
sns.barplot(data=city_product_purchase, x='City', y='Purchase_Amount', hue='Product_Category')
plt.title('Average Purchase Amount by City and Product Category')
plt.xlabel('City')
plt.ylabel('Average Purchase Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
